```python
# Exploratory Data Analysis (EDA)
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Data Visualization
## Visualization 1 : Total Sales by Category
```

```python
category_sales = df.groupby("Category")["Sales"].sum()

plt.figure(figsize=(8,5))
plt.bar(category_sales.index, category_sales.values)

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.grid()

plt.show()
```

```python
### Insight

# 1. Technology category has the highest sales.
# 2. Office Supplies has medium sales.
# 3. Furniture has the lowest sales.
```

```python
## Visualization 2 : Sales by Region
```

```python
region_sales = df.groupby("Region")["Sales"].sum()

plt.figure(figsize=(8,5))
plt.bar(region_sales.index, region_sales.values)

plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.grid()

plt.show()
```

```python
### Insight

# 1. The West region has the highest sales.
# 2. The South region has the lowest sales.
# 3. The company generates more revenue from the West region.
```

```python
## Visualization 3 : Profit by Category
```

```python
category_profit = df.groupby("Category")["Profit"].sum()

plt.figure(figsize=(8,5))
plt.bar(category_profit.index, category_profit.values)

plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Total Profit")
plt.grid()

plt.show()
```

```python
### Insight

# 1. Technology category has the highest profit.
# 2. Furniture has the lowest profit.
# 3. Technology products are more profitable than other categories.
```

```python
## Visualization 4 : Sales Distribution
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["Sales"], bins=20)

plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.grid()

plt.show()
```

```python
### Insight

# 1. Most sales are of small amounts.
# 2. Only a few transactions have very high sales.
# 3. The sales data is positively skewed.
```

```python
## Visualization 5 : Sales vs Profit
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df["Sales"], df["Profit"])

plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.grid()

plt.show()
```

```python
### Insight

# 1. Higher sales generally lead to higher profit.
# 2. Some high sales have low or negative profit due to discounts or costs.
# 3. There is a positive relationship between sales and profit.
```
